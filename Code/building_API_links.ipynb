```python
# Turning list of original menu links to a list
with open("links.txt", "r") as file:
    links = file.read().splitlines()


# Getting school link domains
for link in links:
    domains = [link[8:link.index(".")] for link in links]


#Getting school name from the link
for link in links:
    schools  = [link.split("menu/")[1].split("/lunch")[0] for link in links]


#Building API link using original link, school name, and domain name
api_links=[]
count=-1
for domain in domains:
    count+=1
    api_link = ("https://" + domain + ".api.nutrislice.com/menu/api/weeks/school/" + schools[count] + "/menu-type/lunch/2026/03/01/")
    api_links.append(api_link)


#Saving API links to a text file
with open("api_links", "w") as f:
    f.write("\n".join(api_links))
    
```

```python
# Reopening API Links

with open("api_links", "r") as file:
    api_link_list = file.read().splitlines()
```

```python
urls = api_link_list

# Getting menu ids to build other API Links

import requests
import pandas as pd
from datetime import datetime
import time
menu_ids=[]
for url in urls:
    print(url)
    r = requests.get(url)
    r
    data = r.json()
    menu_ids.append(data.get("menu_type_id"))
```

```python
# Saving menu ids

with open("menu_ids", "w") as f:
    f.write("\n" + str((menu_ids)))
```

```python
# Getting other API links to retrieve school ID

urls = api_link_list
for link in urls:
    school_urls  = [link.split("/weeks/school/")[0] for link in urls]


new_school_urls=[]
for school_url in school_urls:
    school_url = school_url + "/schools/"
    new_school_urls.append(school_url)


with open("api_links", "r") as file:
    api_link_list = file.read().splitlines()
school_slugs=[]
for link in api_link_list:
    slug = link.split("/menu/api/weeks/school/")[1].split("/menu-type/")[0].strip("/")
    school_slugs.append(slug)  
```

```python
# Getting school ids

urls=new_school_urls
import requests
import pandas as pd
from datetime import datetime
import time
school_ids=[]
for url in urls:
    r = requests.get(url)
    r
    data = r.json()
    school_id = next(item["id"] for item in data if item["slug"] == school_slugs[urls.index(url)])
    school_ids.append(school_id)
```

```python
with open("api_links", "r") as file:
    api_link_list = file.read().splitlines()
urls = api_link_list
for link in urls:
    message_urls = [link.split("/weeks/")[0] for link in urls]
```

```python
# Building 

api_links_messages=[]
for link in message_urls:
    link = link + "/messages/school/" + str(school_ids[message_urls.index(link)]) + "/menu-type/" +  str(menu_ids[message_urls.index(link)]) +"/"
    api_links_messages.append(link)
```

```python
with open("api_links_messages", "w") as f:
    f.write("\n".join(api_links_messages))
```
